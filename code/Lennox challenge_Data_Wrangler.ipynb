```python
zip=pd.read_excel('C:\Files and Folders\project\lennox challenge\Final Data\zip_mergedv2.xlsx',0)
```

```python
zip.head()
```

```python
weight=pd.read_excel('C:\Files and Folders\project\lennox challenge\Final Data\zip_mergedv2.xlsx',1)
```

```python
weight.head()
```

```python
merge_zip=weight.merge(zip,left_on='Cust_Zip', right_on='Zip Key')
```

```python
merge_zip.head()
```

```python
for col in merge_zip.columns:
    print(col,merge_zip[col].isna().sum())
```

```python
wsum=lambda x: np.sum(np.multiply(x,merge_zip.loc[x.index,'weight']))
wavg=lambda x: np.average(x,weights=merge_zip.loc[x.index,'weight'])
```

```python
group_zip=merge_zip.groupby(by='Plant').agg({'Total Population': wsum,"Mean Age":wavg , 'Total HH':wsum, 'Median Income HH': wavg, 'Housing Units Current Year Estimates A (2015) Units in Structure (HU) Total: Housing Units': wsum,'LANDSQMI' :wsum,'WATERSQMI':wsum,'POPULATION':wsum,'HSGUNITS':wsum})
```

```python
group_zip.to_csv("C:\Files and Folders\project\lennox challenge\Final Data\zipdata_final.csv")
```
